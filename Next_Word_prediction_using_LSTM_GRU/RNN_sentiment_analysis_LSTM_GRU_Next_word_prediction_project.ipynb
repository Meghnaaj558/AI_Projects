```python
import keras
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
import pandas as pd
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
from keras.utils import pad_sequences
```

```python
X_train=pad_sequences(X_train,padding='post',maxlen=50)
X_test=pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
X_train[1]
```

```python
model=Sequential()
model.add(Embedding(input_dim=10000,output_dim=2,input_length=50))
model.add(SimpleRNN(32,return_sequences=False))
model.add(SimpleRNN(32))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model = keras.models.Sequential()
    # input_dim: size of the vocabulary (number of unique tokens)
    # output_dim: dimension of the embedding vectors
    # input_length: length of the input sequences
model.add(keras.layers.Embedding(input_dim=10000, output_dim=2, input_length=50))
    # Subsequent RNN layers (SimpleRNN, LSTM, GRU) will then process these embeddings
model.add(keras.layers.SimpleRNN(64,return_sequences=False))

model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```

```python
#Data Cleaning
import re
def clean_text(text):
    text=str(text)
    text = text.lower()
    #remove HTML tags
    text = re.sub('<.*?>', '', text)
    #remove special characters
    text = re.sub('[^a-zA-Z]', ' ', text)
    #remove extra whitespace
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
corpus=[]
for sentence in content.split('\n'):
  corpus.append(sentence)
```

```python
corpus1=[]
for i in range(len(corpus)):
  if corpus[i]!='':
    corpus1.append(clean_text(corpus[i]))
corpus1
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(corpus1)
```

```python
corpus=[]

for sentence in corpus1:
  #print(sentence)
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
  #print(tokenized_sentence)
  for i in range(1, len(tokenized_sentence)):
    n_gram_seq = tokenized_sentence[:i+1]
    corpus.append(n_gram_seq)
```

```python
index_to_word = {index: word for word, index in tokenizer.word_index.items()}
```

```python
padded_sequences=pad_sequences(corpus, padding='pre')
padded_sequences.shape
```

```python
X=padded_sequences[:,:-1]
y=padded_sequences[:,-1]
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=8083)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,GRU
```

####LSTM vs GRU

```python
model=Sequential()
model.add(Embedding(input_dim=8083,output_dim=128,input_length=18))
model.add(LSTM(150))
model.add(Dense(8083,activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=50)
```

```python
model1=Sequential()
model1.add(Embedding(input_dim=8083,output_dim=128,input_length=107))
model1.add(GRU(150))
model1.add(Dense(8083,activation='softmax'))
model1.build()
model1.summary()
```

```python
model1.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model1.summary()
```

```python
model1.fit(X,y,epochs=50)
```
